```python
import pandas as pd
```

```python
df = pd.read_csv("swissmetro.dat",'\t')
df = df.reset_index()
df.columns
```

```python
df['QuestionID'] = df['index'] - 9 * (df['ID'] - 1)
df
```

```python
attributes = ['TRAIN_AV',
       'CAR_AV', 'SM_AV', 'TRAIN_TT', 'TRAIN_CO', 'TRAIN_HE', 'SM_TT', 'SM_CO',
       'SM_HE', 'SM_SEATS', 'CAR_TT', 'CAR_CO', 'CHOICE']
socioeco = list(set(df.columns) - set(attributes))
se = df.loc[df['QuestionID'] == 1, socioeco]
se
```

```python
panel = pd.pivot_table(data = df, index = ['ID'], columns = ['QuestionID'], values = attributes)
panel = panel.reset_index()
panel.columns = ['ID'] + ['_'.join((i,f'{j}')) for i, j in panel.columns[1:]]
panel
```

```python
swissmetro = panel.merge(se, on='ID', how='left')
swissmetro.columns
```

```python
swissmetro.to_csv('swissmetro_panel.dat', sep='\t')
```
